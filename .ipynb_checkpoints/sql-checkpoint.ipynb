```python
df = pd.read_excel('narciarstwo.xlsx', index_col=0)
df
```

```python
con = sqlite3.connect('data/wyniki.db')
cur = con.cursor()
df.to_sql('narciarstwo', con, if_exists='replace')
db.to_sql('plywanie', con, if_exists = 'replace')
con.close()
```

```python
con = sqlite3.connect('data/wyniki.db')
dd = pd.read_sql_query("SELECT Kraj, AVG(Punkty) as PKT_sr from narciarstwo GROUP BY Kraj ORDER BY PKT_sr DESC" , con)
con.close()
dd
```

```python
con = sqlite3.connect('data/wyniki.db')
dc = pd.read_sql_query("SELECT Country, count(Country) as Il_zaw from plywanie GROUP BY Country ORDER BY Il_zaw DESC, Country", con)
con.close()
dc
```

```python
con = sqlite3.connect('data/wyniki.db')
cur = con.cursor()
cur.execute("DELETE FROM pracownicy WHERE Nazwisko = 'Burdzinski'")
con.commit()
cur.execute("SELECT * FROM pracownicy")
dd =pd.read_sql_query("SELECT * from pracownicy", con)
con.commit()
con.close()
dd
```
